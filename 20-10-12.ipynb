```python
# 문제 
cond1 = np.array([1,0,1,1,0,0,1], dtype=bool)
cond2 = np.array([0,1,1,1,0,1,1], dtype=bool)
# 두개 다 참이면 0 , cond1 이 참이면 1 , cond2가 참이면 2 , 그외는 3으로 출력
```

```python
for i in range(len(cond1)):
    if(cond1[i] and cond2[i]):
        result.append(0)
    elif cond1[i]:
        result.append(1)
    elif cond2[i]:
        result.append(2)
    else:
        result.append(3)
print("조건문", result)

```

```python
# 중첩된 3항 연산자 
result = np.where(cond1 & cond2 , 0 , np.where(cond1, 1, np.where(cond2,2,3)))
```

```python
import numpy as np 

A = np.array([[1,1,0],[0,1,1],[1,1,1]]) # 3x3 행렬  역행렬은 정방행렬에만 존재
Ainv = np.linalg.inv(A)

# x+ y = 2 , y+z =2  , x+y+z = 3 

b = np.array([[2],[2],[3]])
b 
x = Ainv @ b 
x
```

```python
x , resid , rank ,s = np.linalg.lstsq(A,b) # least square (최소 제곱법 )
print(x)
print(resid)
print(rank)
print(s)
      
```

```python
def fm(p):
    x,y = p
    return np.sin(x)+ 0.25 * x + np.sqrt(y)
x = np.linspace(0,10,20)
y = np.linspace(0,10,20)
X, Y = np.meshgrid(x,y) # 400 개의 정점

Z = fm((X,Y))
x = X.flatten()
y = Y.flatten()
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl 
fig = plt.figure(figsize =(9,6))
ax = fig.gca(projection='3d')

surf = ax.plot_surface(X,Y,Z,rstride =2 , cstride =2 , cmap= mpl.cm.coolwarm, linewidth=0.5, antialiased=True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y')
ax.view_init(elev=60, azim = 0)
fig.colorbar(surf,shrink=0.5 ,aspect=5)
```
